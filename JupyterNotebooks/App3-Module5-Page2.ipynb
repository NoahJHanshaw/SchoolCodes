```python
# Use files.upload( ) to load the data from file of local computer
from google.colab import files
uploaded = files.upload()
my_data = 'ThoraricSurgery.csv'

# Import keras functions to run deep learning
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Import numpy and Tensorflow libraries
import numpy as np
import tensorflow as tf

# Set the seed to produce the same results
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)

# Load the data and then save it in a DataFrame
import os
Data_set = np.loadtxt(my_data, delimiter=",")

# Split Data_set into features (X) and label (Y)
X = Data_set[:,0:17]
Y = Data_set[:,17]

# Eastablish the structure of deep learning
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model (for deep learning)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Train the model with data (X, Y)
model.fit(X, Y, epochs=100, batch_size=10)
```

```python
model.summary()
```

```python
# Input new data (for new patients) into the model to make predicted values
X_new = Data_set[0:10, 0:17]
Y = Data_set[0:10, 17]
Y_prediction = model.predict(X_new)

# Compare 10 predicted values (prediction) with 10 real values (label)
for i in range(10):
    label = Y[i]
    prediction = Y_prediction[i].round(1)
    if prediction>=0.5:
      prediction=1
    else:
      prediction=0
    if label==prediction:
      print("Y = ", label, "\t", "Y_prediction = ", prediction, "\t correct")
    else:
      print("Y = ", label, "\t", "Y_prediction = ", prediction, "\t wrong")

# Display current date/time (CST)
import datetime
import pytz
CST = pytz.timezone('US/Central')
d = datetime.datetime.now(CST)
print("\n\n")
print(d)

# Type your name after Developer: below
print("\nDeveloper: your name here")
```
