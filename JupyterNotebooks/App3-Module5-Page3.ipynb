```python
# import keras and scikit-learn libraries
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
# import numpy, pandas and tensorflow (2.x) libraries
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# import google.colab library to upload the file from local computer
from google.colab import files
uploaded = files.upload()
my_data = 'housing.csv'
```

```python
# set seed to be 0
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)

# Load the data from csv file into df
df = pd.read_csv(my_data, delim_whitespace=True, header=None)

# display the information for df and top four rows in df
print(df.info())
print(df.head())
```

```python
# store the values in df in the dataset
dataset = df.values

# store the data of columns from 0 to 12 in X
# store the data of 13th column in Y
X = dataset[:,0:13]
Y = dataset[:,13]

# Using train_test_split( ) function, split entire dataset into four small datasets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
# Construct the model structure
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))

# Compile the model structure
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# Training the model with datasets (X_train, Y_train)
model.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
# Input X_test into predict() function, in order to get the Y_prediction
# Convert Y_prediction to one-dimensional array using falatten() function
Y_prediction = model.predict(X_test).flatten()

# Compare the top 10 predicted values (prediction) with the top 10 real values (label)
for i in range(10):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print("real price: {:.3f},\tpredicted price: {:.3f}".format(label, prediction))


# Display current date/time (CST)
import datetime
import pytz
CST = pytz.timezone('US/Central')
d = datetime.datetime.now(CST)
print("\n\n")
print(d)

# Type your name after Developer: below
print("\nDeveloper: your name here")
```
